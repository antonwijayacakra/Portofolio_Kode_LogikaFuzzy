```python
#mempersiapkan pustaka  fuzzy yang dibutuhkan 
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
import skfuzzy.membership as mf
```

# FUNGSI KEANGGOTAAN

## Ukuran Layar

```python
#fungsi keanggotaan Input
ukuranlayar = np.arange(0, 21, 0.1)
ukuranlayar_kecil = mf.trapmf(ukuranlayar, [0,0,11,14])
ukuranlayar_sedang = mf.trimf(ukuranlayar, [11,14,17])
ukuranlayar_besar = mf.trapmf(ukuranlayar, [14,17,20,20])
```

```python
#visualisasi fungsi keanggotaan Input ukuran layar
plt.figure(figsize=(10,6))
plt.plot(ukuranlayar, ukuranlayar_kecil, label='Kecil')
plt.plot(ukuranlayar, ukuranlayar_sedang, label='Sedang')
plt.plot(ukuranlayar, ukuranlayar_besar, label='Besar')

plt.xlabel('Ukuran Layar (Inch)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan Fuzzy (scikit-fuzzy)')
plt.legend()
plt.grid(True)
plt.show()
```

## Berat Laptop

```python
berat = np.arange(0, 2601, 1)
berat_ringan = mf.trapmf(berat, [0,0,1800,2200])
berat_sedang = mf.trimf(berat, [1800,2200,2600])
berat_Berat = mf.trapmf(berat, [2200,2600,2600,2600])
```

```python
#visualisasi fungsi keanggotaan Input berat
plt.figure(figsize=(10, 6))
plt.plot(berat, berat_ringan, label='Ringan')
plt.plot(berat, berat_sedang, label='Sedang')
plt.plot(berat, berat_Berat, label='Berat')

plt.xlabel('Berat (gram)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan Fuzzy (scikit-fuzzy)')
plt.legend()
plt.grid(True)
plt.show()
```

## Fungsi Keanggotaan Kecepatan Processor

```python
kecepatan_prosesor = np.arange(0, 5001,1) # dalam MHz

kecepatan_rendah = mf.trapmf(kecepatan_prosesor, [0, 0, 1000, 1700])
kecepatan_sedang = mf.trimf(kecepatan_prosesor, [1600, 1800, 2000])
kecepatan_tinggi = mf.trapmf(kecepatan_prosesor, [1900, 5000, 5000, 5000])
```

```python
#visualisasi fungsi keanggotaan Input prosesr
plt.figure(figsize=(10,6))
plt.plot(kecepatan_prosesor, kecepatan_rendah, label='Rendah')
plt.plot(kecepatan_prosesor, kecepatan_sedang, label='Sedang')
plt.plot(kecepatan_prosesor, kecepatan_tinggi, label='Tinggi')      
plt.xlabel('Kecepatan Prosesor (MHz)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan Fuzzy (scikit-fuzzy)')
plt.legend()
plt.grid(True)
plt.show()
```

## Fungsi keanggotaan Harga Laptop (Output)


```python
harga_laptop = np.arange(0, 15.1, 0.1) # dalam Juta Rupiah
harga_murah = mf.trapmf(harga_laptop, [0, 0, 4, 10])
harga_sedang = mf.trimf(harga_laptop, [4, 10, 15])
harga_mahal = mf.trapmf(harga_laptop, [10, 15, 15, 15])
```

```python
#visualisasi fungsi keanggotaan OUTPUT harga laptop
plt.figure(figsize=(10,6))
plt.plot(harga_laptop, harga_murah, label='Murah')
plt.plot(harga_laptop, harga_sedang, label='Sedang')
plt.plot(harga_laptop, harga_mahal, label='Mahal')      
plt.xlabel('Harga Laptop (Juta Rupiah)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan Fuzzy (scikit-fuzzy)')
plt.legend()
plt.grid(True)
plt.show()
```

# OUTPUT

## NILAI iNPUT dan Ouput

```python
# Nilai Masukan

nilai_ukuranlayar = 13  # contoh input ukuran layar (inch)
nilai_berat =  1500  # contoh input berat (gram)
nilai_kecepatan = 2700  # contoh input kecepatan prosesor (MHz)
```

```python
# Menentukan Nilai Derajat Keanggotaan untuk setiap input
deg_ukuranlayar_kecil = fuzz.interp_membership(ukuranlayar, ukuranlayar_kecil, nilai_ukuranlayar)
deg_ukuranlayar_sedang = fuzz.interp_membership(ukuranlayar, ukuranlayar_sedang, nilai_ukuranlayar)
deg_ukuranlayar_besar = fuzz.interp_membership(ukuranlayar, ukuranlayar_besar, nilai_ukuranlayar)

deg_berat_ringan = fuzz.interp_membership(berat, berat_ringan, nilai_berat)
deg_berat_sedang = fuzz.interp_membership(berat, berat_sedang, nilai_berat)
deg_berat_Berat = fuzz.interp_membership(berat, berat_Berat, nilai_berat)

deg_kecepatan_rendah = fuzz.interp_membership(kecepatan_prosesor, kecepatan_rendah, nilai_kecepatan)
deg_kecepatan_sedang = fuzz.interp_membership(kecepatan_prosesor, kecepatan_sedang, nilai_kecepatan)
deg_kecepatan_tinggi = fuzz.interp_membership(kecepatan_prosesor, kecepatan_tinggi, nilai_kecepatan)

print("Derajat Keanggotaan Ukuran Layar:")
print(f"Kecil: {deg_ukuranlayar_kecil:.4f}")    
print(f"Sedang: {deg_ukuranlayar_sedang:.4f}")
print(f"Besar: {deg_ukuranlayar_besar:.4f}\n")  

print("Derajat Keanggotaan Berat:")
print(f"Ringan: {deg_berat_ringan:.4f}")
print(f"Sedang: {deg_berat_sedang:.4f}")
print(f"Berat: {deg_berat_Berat:.4f}\n")

print("Derajat Keanggotaan Kecepatan Prosesor:")
print(f"Rendah: {deg_kecepatan_rendah:.4f}")
print(f"Sedang: {deg_kecepatan_sedang:.4f}")    
print(f"Tinggi: {deg_kecepatan_tinggi:.4f}")
```

```python
plt.figure(figsize=(8,4))
plt.plot(ukuranlayar, ukuranlayar_kecil, label='Kecil')
plt.plot(ukuranlayar, ukuranlayar_sedang, label='Sedang')
plt.plot(ukuranlayar, ukuranlayar_besar, label='Besar')

plt.axvline(nilai_ukuranlayar, color='k', linestyle='--', label='Input')
plt.scatter([nilai_ukuranlayar]*3,
            [deg_ukuranlayar_kecil, deg_ukuranlayar_sedang, deg_ukuranlayar_besar])

plt.title("Fungsi Keanggotaan Ukuran Layar")
plt.xlabel("Ukuran Layar (Inci)")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(berat, berat_ringan, label='Ringan')
plt.plot(berat, berat_sedang, label='Sedang')
plt.plot(berat, berat_Berat, label='Berat')

plt.axvline(nilai_berat, color='k', linestyle='--', label='Input')
plt.scatter([nilai_berat]*3,
            [deg_berat_ringan, deg_berat_sedang, deg_berat_Berat])

plt.title("Fungsi Keanggotaan Berat Laptop")
plt.xlabel("Berat (Gram)")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(kecepatan_prosesor, kecepatan_rendah, label='Rendah')
plt.plot(kecepatan_prosesor, kecepatan_sedang, label='Sedang')
plt.plot(kecepatan_prosesor, kecepatan_tinggi, label='Tinggi')

plt.axvline(nilai_kecepatan, color='k', linestyle='--', label='Input')
plt.scatter([nilai_kecepatan]*3,
            [deg_kecepatan_rendah, deg_kecepatan_sedang, deg_kecepatan_tinggi])

plt.title("Fungsi Keanggotaan Kecepatan Prosesor")
plt.xlabel("Kecepatan (MHz)")
plt.ylabel("Derajat Keanggotaan")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# FU

# Aturan 1: IF ukuran_layar kecil, berat ringan, kecepatan rendah THEN harga murah
alpha1 = min(deg_ukuranlayar_kecil, deg_berat_ringan, deg_kecepatan_rendah)

# Aturan 2: IF ukuran_layar kecil, berat sedang, kecepatan sedang THEN harga murah
alpha2 = min(deg_ukuranlayar_kecil, deg_berat_sedang, deg_kecepatan_sedang)

# Aturan 3: IF ukuran_layar sedang, berat sedang, kecepatan rendah THEN harga murah
alpha3 = min(deg_ukuranlayar_sedang, deg_berat_sedang, deg_kecepatan_rendah)

# Aturan 4: IF ukuran_layar sedang, berat sedang, kecepatan sedang THEN harga sedang
alpha4 = min(deg_ukuranlayar_sedang, deg_berat_sedang, deg_kecepatan_sedang)

# Aturan 5: IF ukuran_layar sedang, berat ringan, kecepatan sedang THEN harga sedang
alpha5 = min(deg_ukuranlayar_sedang, deg_berat_ringan, deg_kecepatan_sedang)

# Aturan 6: IF ukuran_layar besar, berat berat, kecepatan tinggi THEN harga mahal
alpha6 = min(deg_ukuranlayar_besar, deg_berat_Berat, deg_kecepatan_tinggi)

# Aturan 7: IF ukuran_layar sedang, berat ringan, kecepatan tinggi THEN harga mahal
alpha7 = min(deg_ukuranlayar_sedang, deg_berat_ringan, deg_kecepatan_tinggi)

# Aturan 8: IF ukuran_layar sedang, berat sedang, kecepatan tinggi THEN harga mahal
alpha8 = min(deg_ukuranlayar_sedang, deg_berat_sedang, deg_kecepatan_tinggi)

# Tampilkan nilai firing strength (alpha) untuk setiap aturan
for i, alpha in enumerate([alpha1, alpha2, alpha3, alpha4, alpha5, alpha6, alpha7, alpha8], 1):
    print(f"Alpha Aturan {i}: {alpha:.4f}")
```

```python
#infererensi Fuzzy tsukamoto
# Defuzzifikasi menggunakan metode Tsukamoto
numerator = (alpha1 * 7) + (alpha2 * 7) + (alpha3 * 7) + (alpha4 * 12.5) + (alpha5 * 12.5) + (alpha6 * 12.5) + (alpha7 * 12.5) + (alpha8 * 12.5)
denominator = alpha1 + alpha2 + alpha3 + alpha4 + alpha5 + alpha6 + alpha7 + alpha8
harga_defuzzifikasi = numerator / denominator if denominator != 0 else 0
print(f"\nHarga Laptop (Defuzzifikasi): {harga_defuzzifikasi:.2f} Juta Rupiah")
```

```python
#visualisasi hasil defuzzifikasi (dikasih area warnanya)
plt.figure(figsize=(10,6))  
plt.plot(harga_laptop, harga_murah, label='Murah')
plt.plot(harga_laptop, harga_sedang, label='Sedang')
plt.plot(harga_laptop, harga_mahal, label='Mahal')
plt.axvline(x=harga_defuzzifikasi, color='r', linestyle='--', label='Hasil Defuzzifikasi')
plt.xlabel('Harga Laptop (Juta Rupiah)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Hasil Defuzzifikasi Harga Laptop')
plt.legend()
plt.grid(True)
```
