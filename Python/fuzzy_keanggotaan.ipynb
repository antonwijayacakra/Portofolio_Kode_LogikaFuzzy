```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
import skfuzzy.membership as mf
```

*TENTUKAN VARIABEL  DAHULU*

```python
suhu  = np.arange(0, 101, 1)   # Rentang suhu dari 0 hingga 100 dengan step 1
#np.arange(start, stop, step)
print (suhu) # pada dasarnya membuat data ARRAY
```

# Keanggotaan segitiga 

fuzz.trimf(x, [a, b, c]) ;
  a keanggotaan 0 ;
  b keanggotaan 1 ;
  c keanggotaan 0 ;

```python
suhu_dingin = mf.trimf(suhu, [0, 0, 50]) # awal: 0; puncak= 0; akhir = 50
```

```python
suhu_sedang = mf.trimf(suhu, [30, 50, 80]) 
```

```python
suhu_tinggi = mf.trimf(suhu,[50,100,100] )
```

```python
#Visulisasi kan
plt.figure(figsize=(8, 5))
plt.plot(suhu, suhu_dingin, label="Dingin", color='blue')
plt.plot(suhu, suhu_sedang, label="sedang", color='red')
plt.plot(suhu, suhu_tinggi, label="sedang", color='yellow')
plt.title("Fungsi Keanggotaan Suhu")
plt.xlabel("Suhu (Â°C)")
plt.ylabel("Derajat Keanggotaan")
plt.grid()
plt.show()
```

## Titik nya pada fungsi trimf

```python
# Menampilkan grafik fungsi keanggotaan
plt.figure(figsize=(8, 5))

# Titik-titik penting
titik_x_dingin = [10, 25, 50]
titik_y_dingin = [0, 1, 0]


# Menampilkan titik-titik penting dengan scatter
plt.scatter(titik_x_dingin, titik_y_dingin, color='red', zorder=3)


# Label pada titik-titik penting
for i, txt in enumerate(zip(titik_x_dingin, titik_y_dingin)):
    plt.annotate(f"({txt[0]}, {txt[1]})", (txt[0], txt[1]), textcoords="offset points", xytext=(5,5), ha='right')



plt.title("Fungsi Keanggotaan Segitiga - Suhu")
plt.xlabel("Suhu (Â°C)")
plt.ylabel("Derajat Keanggotaan")
plt.grid()
plt.show()
```

# Keanggotaan trapesium 


[a, b, c, d] â Titik-titik trapesium: a â Titik awal (keanggotaan 0);
b â Titik mulai naik (keanggotaan naik dari 0 ke 1);
c â Titik mulai turun (masih keanggotaan 1);
d â Titik akhir (keanggotaan kembali ke 0);


```python
suhu_dingin_TR = mf.trapmf(suhu, [0, 20, 50, 70])
```
